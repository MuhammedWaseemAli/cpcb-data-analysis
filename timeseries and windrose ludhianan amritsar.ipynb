```python
import pandas as pd
import matplotlib.pyplot as plt
from windrose import WindroseAxes

# Load the CSV file
file_path = r"C:\Users\wasee\Downloads\Raw_data_15Min_2023_site_253_Punjab_Agricultural_University_Ludhiana_PPCB_15Min.csv"
df = pd.read_csv(file_path)

# Convert the Timestamp column to datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%Y-%m-%d %H:%M:%S')

# Filter the data for March 23 and 24
df_march_23_24 = df[(df['Timestamp'].dt.date == pd.Timestamp('2023-03-23').date()) |
                    (df['Timestamp'].dt.date == pd.Timestamp('2023-03-24').date())]

# Plot the wind rose
ws = df_march_23_24['WS (m/s)']
wd = df_march_23_24['WD (deg)']

fig = plt.figure(figsize=(8, 8))
ax = WindroseAxes.from_ax()
ax.bar(wd, ws, normed=True, opening=0.8, edgecolor='white')
ax.set_legend()
plt.title('Wind Rose for March 23 and 24')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from windrose import WindroseAxes

# Load the CSV file
file_path = r"C:\Users\wasee\Downloads\Raw_data_15Min_2023_site_253_Punjab_Agricultural_University_Ludhiana_PPCB_15Min.csv"
df = pd.read_csv(file_path)

# Convert the Timestamp column to datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%Y-%m-%d %H:%M:%S')

# Filter the data for the month of November
df_november = df[df['Timestamp'].dt.month == 11]

# Separate data into nighttime (20:00 to 06:00) and daytime (06:00 to 20:00)
nighttime = df_november[(df_november['Timestamp'].dt.hour >= 20) | (df_november['Timestamp'].dt.hour < 6)]
daytime = df_november[(df_november['Timestamp'].dt.hour >= 6) & (df_november['Timestamp'].dt.hour < 20)]

# Extract wind speed and wind direction for nighttime and daytime
ws_night = nighttime['WS (m/s)']
wd_night = nighttime['WD (deg)']
ws_day = daytime['WS (m/s)']
wd_day = daytime['WD (deg)']

# Plot the wind rose for nighttime
fig = plt.figure(figsize=(16, 8))

ax_night = fig.add_subplot(121, projection="windrose")
ax_night.bar(wd_night, ws_night, normed=True, opening=0.8, edgecolor='white')
ax_night.set_legend()
ax_night.set_title('Wind Rose for Nighttime (20:00 to 06:00) in November')

# Plot the wind rose for daytime
ax_day = fig.add_subplot(122, projection="windrose")
ax_day.bar(wd_day, ws_day, normed=True, opening=0.8, edgecolor='white')
ax_day.set_legend()
ax_day.set_title('Wind Rose for Daytime (06:00 to 20:00) in November')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from windrose import WindroseAxes
from tabulate import tabulate

# Load the CSV file
file_path = r"C:\Users\wasee\Downloads\Raw_data_15Min_2023_site_253_Punjab_Agricultural_University_Ludhiana_PPCB_15Min.csv"
df = pd.read_csv(file_path)

# Convert the Timestamp column to datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%Y-%m-%d %H:%M:%S')

# Set the Timestamp as the index
df.set_index('Timestamp', inplace=True)

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['PM10 (g/m)'], label='PM10 (g/m)', color='blue')
plt.xlabel('Timestamp')
plt.ylabel('PM10 (g/m)')
plt.title('Time vs PM10')
plt.legend()
plt.grid(True)
plt.show()

# Find the top 100 timestamps with the highest PM10 concentrations
top_100_pm10 = df['PM10 (g/m)'].nlargest(100)

# Print the top 100 PM10 concentrations
print("Top 100 PM10 Concentrations:")
top_100_pm10_tabulated = top_100_pm10.reset_index()
print(tabulate(top_100_pm10_tabulated, headers=['Timestamp', 'PM10 (g/m)'], tablefmt='grid'))

# Filter the dataframe to include only the rows corresponding to the top 100 PM10 concentrations
top_100_timestamps = top_100_pm10_tabulated['Timestamp']
df_top_100 = df.loc[top_100_timestamps]

# Extract wind speed and wind direction for the top 100 PM10 concentration dates
ws_top_100 = df_top_100['WS (m/s)']
wd_top_100 = df_top_100['WD (deg)']

# Plot the wind rose for the top 100 PM10 concentration dates
fig = plt.figure(figsize=(8, 8))
ax = WindroseAxes.from_ax()
ax.bar(wd_top_100, ws_top_100, normed=True, opening=0.8, edgecolor='white')
ax.set_legend()
plt.title('Wind Rose for Dates with Top 100 PM10 Concentrations')
plt.show()
```
